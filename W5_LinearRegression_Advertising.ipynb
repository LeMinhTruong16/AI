```python
from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'mtcars.csv'
data = pd.read_csv(url)
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])
score = model.score(X=pd.DataFrame(data['wt']), y=data['mpg'])
print("do chinh xac", score)

y_pred = model.predict(X=pd.DataFrame(data['wt']))
residuals = data['mpg'] - y_pred

residuals.describe()

data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
plt.plot(data['wt'],y_pred,color='red')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'mtcars.csv'
data = pd.read_csv(url)

degree = 3
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X=pd.DataFrame(data['wt']))
model = LinearRegression()
model.fit(X=X_poly, y=data['mpg'])

vange = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame({'wt': vange})
predictor_poly = poly_features.transform(predictor)

y_pred = model.predict(X=predictor_poly)

data.plot(kind='scatter', x='wt', y='mpg', figsize=(9,9), color='black', xlim=(0, 7))
plt.plot(vange, y_pred, color='red')
plt.show()
print(' chnh xc ca m hnh')
     
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**2,data['wt']**3,data['wt']**4]).T
y =data['mpg']
model.fit(X,y=data['mpg'])

print('Intercept cua mo hinh bac 4: ')
print(model.intercept_)
print('Coeffient cua mo hinh bac 4: ')
print(model.coef_)
print('Do chinh xac cua mo hinh')
print(model.score(X,y))

range=np.arange(1.5,5.5,0.1)
predictor = pd.DataFrame([range,range**2,range**3,range**4]).T
y_pred = model.predict(X=predictor)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='green')
```
